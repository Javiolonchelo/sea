# Filtros de cruce

```python
import numpy as np
from sympy.abc import s
from sympy.physics.control.lti import TransferFunction
from sympy.physics.control.control_plots import bode_plot
import math

def my_bode_plot(tf, normalized=True):
    if normalized:
        bode_plot(tf, initial_exp=-2, final_exp=1, show_axes=True)
    else:
        bode_plot(tf, initial_exp=1, final_exp=5, show_axes=True)


# Def the transfer function as a symbolic expression
w = np.logspace(2*math.pi*20, 2*math.pi*20000, 1000)
wl = 2*math.pi * 1800
Qts = 10
Ts = 1/1
Hl = 1 / (s**2 *Ts + s*Ts/Qts + 1)
Hh = -s**3 / (s**3 + 2*s**2 + 2*s + 1)
# Hl = Hl.subs(s, s/wl)
# Hh = Hh.subs(s, s/wl)
num_Hl, den_Hl = Hl.as_numer_denom()
num_Hh, den_Hh = Hh.as_numer_denom()

tf_Hl = TransferFunction(num_Hl, den_Hl, s)
tf_Hh = TransferFunction(num_Hh, den_Hh, s)

display(num_Hl/den_Hl, num_Hh/den_Hh)

# Bode plot


my_bode_plot(tf_Hl)
my_bode_plot(tf_Hh)
```

```python
H_total = num_Hl/den_Hl + num_Hh/den_Hh


display("H total", H_total.simplify())
# H_total = H_total.subs(s, s/wl)
display("H total at wl", H_total)

# get num and den
num, den = H_total.as_numer_denom()
tf_total = TransferFunction(num, den, s)
my_bode_plot(tf_total)
```
